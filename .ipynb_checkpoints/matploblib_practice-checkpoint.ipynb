```python
#Import dependencies
import matplotlib.pyplot as plt
```

```python
#Set x-axis to a list of strings for each month
x_axis = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

#Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis)
plt.xlabel("Date")
plt.ylabel("Savings($)")
plt.title('Savings Account Results')
plt.show()
```

```python


#Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston',)
#Create labels for the x and yaxes 
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Set the y limit between 0 and 45
plt.ylim(0, 45)
# Create a title
plt.title("PyBer Fare by Month")
#Add a grid
plt.grid()
# Add the legend
plt.legend()

#if you want to change 
#the figure size to 20 pixels wide by 10 pixels high do plt.figure(figsize=(20,10)) or 
#plt.subplots(figsize=(20,10))
```

```python
#set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

#set the y-axis to a list of floats at the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09 ]
```

```python
#Create the plot
plt.bar(x_axis, y_axis)
```

```python
#Create the plot
plt.bar(x_axis, y_axis, color="green", label='Boston')

#Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Create a title
plt.title("PyBer Fare by Month")

#Add the legend
plt.legend()
```

```python
#Lets create a horizontal bar chart to display the music groups among teens, number of votes by county
plt.barh(x_axis, y_axis)
```

```python
#If we want the data on opposite axes, switch the arguments in the barh() function  but this no good
plt.barh(y_axis, x_axis)
```

```python
#Lets switch back because that is the best way to represent the data to convey each months average fare
#Create the plot to have y-axis to have the months in ascending order, using gca() method to get current axes
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
#Create the plot
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
plt.gca().invert_yaxis()
#Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Create a title
plt.title("PyBer Fare by Month")

#Add the legend
plt.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="k", label='New York')
plt.xlabel("Date")
plt.ylabel("Fare($)")
```

```python
#Create a horizontal bar chart using object-oriented interface method (ax.barh) with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
#Create the plot with ax.plt() to switch arguments in the barh() function
fig, ax = plt.subplots()
ax.barh(y_axis, x_axis)
```

```python
fig, ax = plt.subplots()
# Set the x-axis to a list of strings for each month.
y_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
x_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

ax.barh(y_axis, x_axis, color="cyan", label ='Chicago')
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.gca().invert_yaxis()
plt.title("PyBer Fare by Month")
plt.legend()
```
